```python
class library:
    def __init__(self):
        self.books = {
        'The Great Gatsby': 'None',
        'Anna Karenina': 'None',
        'One Hundred Days of Solitude': 'None'
        }
    
    def display_book(self):
        print("The available books are:")
        print(self.books)
        #for x in self.books:
          #print(x)
        
    def add_book(self):
        name=input("Name of the book?: ")
        self.books.update({name:"None"})
        print("Thanks for donating!")
        
    def lend_book(self):
        flag=0
        name=input("Enter your name?: ")
        demand=input("Which book would you like to have?: ")
        if self.books[demand] == "None":
            print("You can have it!")
            flag = 1
            self.books[demand]=name
        else:
            print("Sorry, not available!")
    
    def return_book(self):
        ret=input("Which book would you like to return?: ")
        if self.books[ret]=="None":
            print("Wrong Input")
        else:
            self.books[ret]="None"
            print("Book successfully returned")
        
obj=library()
        
choice=1
while choice!=0:
    
    print("1. Display Book")
    print("2. Add Book")
    print("3. Lend Book")
    print("4. Return Book")
    print("5. Exit")
    choice=int(input("Enter choice: "))
    if choice==1:
        obj.display_book()
    elif choice==2:
        obj.add_book()
    elif choice==3:
        obj.lend_book()
    elif choice==4:
        obj.return_book()
    elif choice==5:
        print("Exiting!")
        break
    else:
        print("Invalid choice!!")
```
